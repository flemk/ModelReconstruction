# Generate a time series

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def f(t,y): return -a[0]*y**2+a[1]*y**4
def g(t,y): return a[0]*y
```

```python
a=[1.,1.]
y0=[0.01]
tspan=(0.,10.)
teval=np.arange(0.,10.,0.1)
sol=solve_ivp(g,tspan,y0,method='RK45',t_eval=teval,args=a)
```

```python
%matplotlib notebook
plt.plot(sol.y[0])
```
